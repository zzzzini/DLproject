```python
def get_sampling_rate(wav_file):
  sound = AudioSegment.from_wav(wav_file)
  return sound.frame_rate
```

```python
wav_file = '/content/drive/MyDrive/newwav/AU01-0000.wav'

# 샘플링 주파수 얻기
sampling_rate = get_sampling_rate(wav_file)
print(f'WAV 파일의 샘플링 주파수: {sampling_rate}Hz')
```

```python
# 스테레오 타입을 모노타입으로 바꾸고, 비트 전송률 바꾸는 코드

from pydub import AudioSegment
import os

# 변환할 파일이 있는 폴더와 결과를 저장할 폴더
source_folder = '/content/drive/MyDrive/newwav'
output_folder = '/content/drive/MyDrive/datav2'

# 소스 폴더에서 모든 wav 파일을 찾아 처리
for file_name in os.listdir(source_folder):
    if file_name.endswith('.wav'):
        # 파일 경로 생성
        source_path = os.path.join(source_folder, file_name)
        output_path = os.path.join(output_folder, file_name)

        # 오디오 파일 로드
        audio = AudioSegment.from_wav(source_path)

        # 스테레오에서 모노로 변환
        mono_audio = audio.set_channels(1)

        # 비트 전송률을 352kbps로 설정하여 파일 저장
        mono_audio.export(output_path, format="wav", bitrate="352k")

print("변환 완료")
```
