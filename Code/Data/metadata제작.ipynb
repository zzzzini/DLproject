```python
# 파일 이름 생성 함수
def generate_filenames(string):
    filenames = []
    for i in range(string,string+1):
        filenames.append(f"DT01-{str(i).zfill(4)}.wav")
    return filenames

meta.loc[:,0]=meta.loc[:,0].apply(generate_filenames)
```

```python
import inflect

def convert_numbers_to_words(text):
    # inflect 라이브러리를 사용하여 숫자를 문자열로 변환
    p = inflect.engine()
    words = text.split()
    for i, word in enumerate(words):
        # word가 숫자인지 확인
        if word.isdigit():
            # 숫자를 문자열로 변환하여 대체
            words[i] = p.number_to_words(word)
    # 리스트를 다시 문장으로 결합
    return ' '.join(words)

def replace_dollars(string):
    return string.replace('$', 'dollars')

# meta.loc[:, 2]에 있는 각 행의 문장에 함수 적용
meta.loc[:, 2] = meta.loc[:, 2].apply(convert_numbers_to_words)

meta.loc[:, 2] = meta.loc[:, 2].apply(replace_dollars)


print(meta.loc[:, 2])
```
