```python
from pydub import AudioSegment, silence

def extract_silence_from_audio(audio_path, output_folder, min_silence_len=1000, silence_thresh=-40):
    # 음성 파일 불러오기
    audio = AudioSegment.from_file(audio_path)

    # 무음 구간 찾기
    silences = silence.detect_silence(audio, min_silence_len=min_silence_len, silence_thresh=silence_thresh)

    # 무음 구간을 시간으로 변환 (PyDub은 밀리초 단위를 사용)
    silences = [(start / 1000.0, end / 1000.0) for start, end in silences]

    # 무음 구간만을 별도의 파일로 저장
    for i, (start, end) in enumerate(silences, start=1):
        silent_segment = audio[start * 1000:end * 1000]  # 밀리초 단위로 인덱싱
        output_path = f"{output_folder}/silence_{i}.wav"
        silent_segment.export(output_path, format="wav")
        print(f"Silence segment {i} exported: {output_path}")

# 사용 예
audio_path = "/content/drive/MyDrive/02_asac_04_dl_1jo/silence/audio_Hz/DT001-0001.wav"
output_folder = "/content/drive/MyDrive/02_asac_04_dl_1jo/silence"
extract_silence_from_audio(audio_path, output_folder)
```
